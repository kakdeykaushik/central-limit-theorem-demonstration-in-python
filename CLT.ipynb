```python
# imports
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
```

## Defining Central Limit Theoram class

```python
class CLT:

  def __init__(self, data, sample_size=(500, 10)):
    self.data = data
    self.sample_size = sample_size

    self.plot_graph(self.data, title="Original Distribution")
    self.calc_mean()
    self.plot_graph(self.means, title="CLT Distribution")



  def calc_mean(self):
    sample = np.random.choice(self.data, size=self.sample_size, replace=True)
    self.means = sample.mean(axis=1)




  def plot_graph(self, data, bins=None, title=""):
    plt.figure(figsize=(12, 8))
    plt.title(title, fontsize=16, loc='left', fontweight="bold")
    plt.title("by Kaushik Kakdey", loc='right', fontsize=8, color='grey', style='italic')

    sns.histplot(data, kde=True, bins=bins if bins else 'auto', color='orange', ec='white', stat="density");
    sns.kdeplot(data, color="black")
```
